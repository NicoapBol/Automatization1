```python
#Librerias
import pandas as pd
import numpy as np
import tabulate 
from tabulate import tabulate
```

```python
#Carga de la base plana en archivo csv.
error_bad_lines=False
bpl = pd.read_csv(r'SELLOUT_D.CSV', sep=";", decimal= ',', dtype= {'Quantidade': 'float64'}, low_memory=False)
```

```python
#Test
#bpl.head
```

```python
#Seleccionar solo los valores Ven y Bon
bp = bpl.loc[((bpl["Tipo Operação"] =="VEN") | (bpl["Tipo Operação"] == "BON"))]
```

```python
#Test 2
#print(bp)
```

```python
#Seleccionar ventas con cantidad >0
bp_1 = bp.loc[(bp['Quantidade'] >0)]
```

```python
#Copiado e insertar columna "vendedor pedido" y separar
copy_1 = bp_1['Vendedor Pedido']
bp_1.insert(loc = 5, column = 'Ven_Pedido', value=copy_1)
bp_1[['Cod Ven','Nombre Vendedor']] = bp_1['Ven_Pedido'].str.split('-', expand=True)   
```

```python
#Copiado e insertar columna "produto" y separar
copy_1 = bp_1['Produto']
bp_1.insert(loc = 5, column = 'Product', value=copy_1)
bp_1[['Cod Pro','Nombre Pro']] = bp_1['Product'].str.split('-', expand=True)  

#Generar columna para calibre
copy_1 = bp_1['Produto']
bp_1.insert(loc = 5, column = 'Cal', value=copy_1)
bp_1[['Borrar','Calibre']] = bp_1['Product'].str.split('X', expand=True) 
```

```python
#Copiado e insertar columna "Grupo" y separar
copy_1 = bp_1['Grupo']
bp_1.insert(loc = 5, column = 'Grup', value=copy_1)
bp_1[['Cod Gru','Nombre Gru']] = bp_1['Grup'].str.split('-', expand=True)  
```

```python
#Copiado e insertar columna "Litro" y convertir a hectolitros
copy_1 = bp_1['Litro']
bp_1.insert(loc = 14, column = 'Hectos', value=copy_1/100)
```
